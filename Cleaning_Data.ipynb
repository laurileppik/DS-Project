```python
import pandas as pd

data=pd.read_csv("original_TMDB_movie_dataset_v11.csv")
```

```python
data.columns
```

```python
filtered_data = data[data['vote_count'] >= 50]

print(filtered_data)
```

```python
#This cell is used for testing different column values for cleaning the dataset (duplicate titles removed, status removed,
#all revenues with less than 25 000 removed/replaced, 0 runtime movies removed, all budgets with less than 25 000 removed/replaced)
def column_values(test_column,data):
    print(data.columns)
    unique_values = data[test_column].unique()
    print(unique_values)
    print("Nan values: " + str(data[test_column].isna().sum()))

    # Get the count of each unique value
    sorted_value_counts = data[test_column].value_counts().sort_index(ascending=False)
    print(sorted_value_counts)
    
def column_values_for_release_date(test_column,data):
    print(data.columns)
    unique_values = data[test_column].unique()
    print(unique_values)
    
    df=data.copy()
     # Convert the release_date column to datetime
    df[test_column] = pd.to_datetime(df[test_column])

    # Extract the decade from the release_date
    df['decade'] = (df[test_column].dt.year//10)*10
    # Get the count of each unique value
    sorted_value_counts = df['decade'].value_counts().sort_index(ascending=False)
    print(sorted_value_counts)
    
column_values("spoken_languages",filtered_data)
#column_values_for_release_date("release_date",filtered_data)
```

```python
return_rows = filtered_data[filtered_data['title'] == 'Return']

# Print the rows
print("Rows with the title 'Return':")
print(return_rows)
```

```python
#Remove all duplicate titles from dataset
column_values("title",filtered_data)
print(filtered_data.shape)
filtered_data=filtered_data.drop_duplicates(subset='title', keep='first')
print(filtered_data.shape)
```

```python
#Remove status since it is unimportant all of the movies are already out in this dataset.
column_values("status",filtered_data)
filtered_data=filtered_data.drop("status",axis=1)
filtered_data.columns
```

```python
#Remove movies with runtime of 0
column_values("runtime",filtered_data)
print(filtered_data.shape)
filtered_data = filtered_data[filtered_data['runtime']>0]
print(filtered_data.shape)
```

```python
#Replace revenues for movies that have a revenue of less than 10000 with median value of the revenues
column_values("revenue",filtered_data)
filtered_data_by_median=filtered_data.copy()

median_revenue = filtered_data[filtered_data['revenue'] >= 10000]['revenue'].median()
filtered_data_by_median.loc[filtered_data_by_median['revenue'] < 10000, 'revenue'] = median_revenue

column_values("revenue", filtered_data_by_median)
print(filtered_data_by_median['revenue'].value_counts())
print(filtered_data_by_median.shape)
```

```python
#Replace revenues for movies that have a budget of less than 10000 with median value of the budgets (using already filtered revenue)
column_values("budget",filtered_data_by_median)
filtered_data_by_budget=filtered_data_by_median.copy()

median_budget = filtered_data_by_median[filtered_data_by_median['budget'] >= 10000]['budget'].median()
filtered_data_by_budget.loc[filtered_data_by_budget['budget'] < 10000, 'budget'] = median_budget

column_values("budget", filtered_data_by_budget)
print(filtered_data_by_budget['budget'].value_counts())
print(filtered_data_by_budget.shape)
```

```python
#Remove movies that have a revenue less than 10000
column_values("revenue",filtered_data)
print(filtered_data.shape)
filtered_data = filtered_data[filtered_data['revenue'] >= 10000]
print(filtered_data.shape)
```

```python
#Remove movies that have a budget less than 10000
column_values("budget",filtered_data)
print(filtered_data.shape)
filtered_data = filtered_data[filtered_data['budget'] >= 10000]
print(filtered_data.shape)
```

```python
filtered_data.to_csv("fil_data_movies_lessthan10k_removed.csv", index=False)
filtered_data_by_budget.to_csv("fil_data_movies_lessthan10k_replaced.csv", index=False)
```
